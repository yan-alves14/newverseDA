## **Lucro bruto ao longo da companhia**

```python
import pandas as pd

import matplotlib.pyplot as plt
#Configura a biblioteca para a plotagem em grafico
%matplotlib inline 


dados = pd.read_csv("dados_de_lucro_bruto.csv", sep=";")
#barra_tiju = (aluguel_residenciais[aluguel_residenciais["Bairro"] == "Barra da Tijuca"])

dados

data = dados["Mes"].unique()

data

valor=dados["Lucro Bruto"]
valor

fig, ax = plt.subplots(figsize=(12,8))

#Cores das barras do grafico
cor=["cyan","red","black", "orange", "green","purple", "gold", "blue"]

ax.bar(data, valor, color=cor, edgecolor="black", width=0.51)
ax.set_title("Lucro Bruto ao longo da Compania", fontsize=20)
ax.set_xlabel("meses", fontsize=15)
ax.set_ylabel("Lucro Bruto", fontsize=15)

#Exibe as media de preço dos alugueis nas barras
for data, valor, in enumerate(valor):
  plt.text(x=data-.64,y=valor+18822,s=f"R$ {valor:.2f}", fontdict=dict(fontsize=8))
plt.tight_layout()
plt.show()
```

## **Acesso dos alunos ao cursos**

```python

import pandas as pd

import matplotlib.pyplot as plt
#Configura a biblioteca para a plotagem em grafico
%matplotlib inline 


dados = pd.read_csv("Cursos_acessos.csv", sep=";")
#barra_tiju = (aluguel_residenciais[aluguel_residenciais["Bairro"] == "Barra da Tijuca"])

dados

data = dados["mes"].unique()

data

valor=dados["acessos"]
valor


#----------------------------------

fig, ax = plt.subplots(figsize=(12,8))

#Cores das barras do grafico
cor=["purple", "gold", "blue","cyan","red","black", "orange", "green","green"]

ax.bar(data, valor, color=cor, edgecolor="black", width=0.51)
ax.set_title("Alunos que acessaram os cursos aos meses", fontsize=20)
ax.set_xlabel("meses", fontsize=18)
ax.set_ylabel("acessos", fontsize=18)

#Exibe as media de preço dos alugueis nas barras
for data, valor, in enumerate(valor):
  plt.text(x=data-.20,y=valor+182,s=f"{valor:.0f}", fontdict=dict(fontsize=16))
plt.tight_layout()
plt.show()
```
